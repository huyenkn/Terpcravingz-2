```python
import pandas as pd
import numpy as np
import re
import regex
from pprint import pprint
import string
import nltk
```

```python
df = pd.read_csv("Review.csv")
```

```python
df1 = pd.DataFrame(df)
df1['Type'] = ""
```

```python
df1.head()
```

```python
positive = open("Positive.txt","r")
negative = open("Negative.txt","r")
```

```python
list_pos = []
for x in positive:
    list_pos.append(x)
```

```python
list_neg = []
for x in negative:
    list_neg.append(x)
```

```python
for i in range(len(list_pos)):
    list_pos[i] = list_pos[i].rstrip("\n\r")
    list_pos[i] = list_pos[i].lower()
```

```python
for i in range(len(list_neg)):
    list_neg[i] = list_neg[i].rstrip("\n\r")
    list_neg[i] = list_neg[i].lower()
```

```python
list_pos
```

```python
list_neg
```

```python
translator = str.maketrans('', '', string.punctuation)

listit=[]


for i in df1['revText']:
    i = str(i)
    senti = i.lower()
    senti = senti.translate(translator) # remove punctuation
    senti = senti.encode('ascii', 'ignore').decode("utf-8") # remove unicode
    senti = re.sub(r"http\S+", "", senti) # remove url
    bagofwordsi = nltk.word_tokenize(senti) #collect list of words
    
    
    dicti = {}
    for j in (bagofwordsi):
        #print(i)
        if j in list_pos:
            dicti[j] = "Positive"
        elif j in list_neg:
            dicti[j] = "Negative"
            
    #print(dicti)     
    pos_count = 0
    neg_count = 0
    for k in dicti.values():
        if k == "Positive":
            pos_count = pos_count +1
        if k == "Negative":
            neg_count = neg_count +1
            
    totalcount = 0
    totalcount = pos_count + neg_count
    if totalcount != 0 :
        pos_percent = pos_count / totalcount
        #print(pos_percent)
        neg_percent = neg_count / totalcount
        #print(neg_percent)
        if pos_percent > neg_percent:
            listit.append("Positive")
        elif pos_percent < neg_percent:
            listit.append("Negative")
        else: 
            listit.append("Neutral")
    else:
        listit.append("Neutral")
    
df1['Type'] = listit    
    
```
